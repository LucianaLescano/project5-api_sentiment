Getting information from our database

```python
import pandas as pd
from pymongo import MongoClient
import requests
import json
```

### First request
#### Getting all messages from a certain character from IV collection

```python
url_query_iv = 'http://127.0.0.1:5000/IV/'
```

```python
wanted_character = 'LUKE' 
```

```python
all_dialogues = requests.get(url_query_iv + wanted_character)
all_dialogues = all_dialogues.json()
all_dialogues
```
